```python
import datetime

def display_menu():

    print("\nInventory Management System")

    print("1. Add item")

    print("2. Remove item")

    print("3. Display total number of items")

    print("4. Search for item")

    print("5. Display last update date and time")

    print("6. Exit")
```

```python
# Creating a dictionary to store the data of name and quantity as key and avlue
Inventory={}
#Function to accept the item name and quantity from the user and add it to dictionary
def add_item():
    name=input("ENter Item name")
    quantity=int(input("Enter Quantity"))
    # To check if the entered quantity is 0 or not
    if quantity<0:
        print("Quantity must be more than 0")
        return
    if name in Inventory:
        # Increase the quantity od the item if the name is updated as the key
        Inventory[name]+=quantity
    else:
        Inventory[name]=quantity
    # Print the data of name and quantity with teh date when it is added
    print(f"{quantity} {name}s are added in the inventory on {datetime.datetime.now()} ")
    
```

```python
# Function to remove the data form the dictionary
def remove_item():
    name=input("ENter Item name")
    quantity=int(input("Enter Quantity"))
    # Check the quantity of an item that should not be 0
    if quantity<0:
        print("Quantity must be more than 0")
        return
    if name in Inventory:
        if Inventory[name]>=quantity:
            # If the entered quantity is less than that in inventory the subtracting it from the inventory
            Inventory[name]-=quantity
    #    Displaying the name and quantity of the data that has been removed from inventory with the date updated.
        print(f"{quantity} {name} is removed from the inventory on {datetime.datetime.now()}")
            
```

```python
def total_inventory():
    total = sum(Inventory.values())
    print(f"Total number of items in inventory: {total} till {datetime.datetime.now()}")
```

```python
# Function to check the total quantity in the inventory
def search_item():
    name = input("Enter item name to search: ")
    # Checking if the data is in the inventroy or it is empty
    if name in Inventory:
        #Displaying the data in the inventory
        print(f"{name}: {Inventory[name]} Updated on {datetime.datetime.now()}")
    else:
        print("Item not found in inventory.")
```

```python
def display_last_update():
    update = max(Inventory.items(), key=lambda x: x[1])
    if update[0]:
        print(f"Last update date and time: {datetime.datetime.now()}")
    else:
        print("Inventory is empty.")
```

```python
def main():
    # Creating the menu to select the option to run funvtion 
    while True:
        display_menu()
        choice = input("Enter your choice (1-6): ")
        #TO add the item in the inventory press 1
        if choice == '1':
            add_item()
        #To remove the item from the inventory press 2
        elif choice == '2':
            remove_item()
        # To view all items in the inventory, press 3.
        elif choice == '3':
            total_inventory()
        elif choice=='4':
        # To search the item in the inventory, press 4.
            search_item()
        # To check the last updated time, press 5.
        elif choice=='5':
            display_last_update()
        elif choice == '6':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 6.")

if __name__ == "__main__":
    main()
```
