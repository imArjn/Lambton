```python
import random
#Creating a class Time having attributes hour, minute and seconds
class Time:
    #Method having constructor hr,min and sec.
    def __init__(self, hr=0, min=0, sec=0):
        self.hr = hr
        self.min = min
        self.sec = sec
    #Method to format the value in the string format
    def __str__(self):
        return "{:02d}:{:02d}:{:02d}".format(self.hr, self.min, self.sec)
    # Method ti add time object
    def __add__(self, other):
        seconds = self.hr * 3600 + self.min * 60 + self.sec
        seconds += other.hr * 3600 + other.min * 60 + other.sec
        
        return Time(seconds // 3600 % 24, seconds // 60 % 60, seconds % 60)
    # Method to subtract time object
    def __sub__(self, other):
        seconds = self.hr * 3600 + self.min * 60 + self.sec
        seconds -= other.hr * 3600 + other.min * 60 + other.sec
        
        return Time(seconds // 3600 % 24, seconds // 60 % 60, seconds % 60)

```

```python
# Creating a function to perform operaition on time
def main(h1, m1, s1, h2, m2, s2):
    #Storing the value of first time in first variable
    first = Time(h1, m1, s1)
    # Storing the value of second variable in second variable
    second = Time(h2, m2, s2)

    print("First:", first)
    print("Second:", second)
    print("First + Second:", first+second)
    print("First - Second:", first-second)
    print()

# TO generate random numbr between 0-24 for hour
def h():
    return random.randint(0, 23)

# To generate random number between 0-59 for minute
def m():
    return random.randint(0, 59)

# To generate random number between 0-59 for seconds
def s():
    return random.randint(0, 59)

if __name__ == "__main__":
    for _ in range(10):
        main(h(), m(), s(), h(), m(), s())
```
