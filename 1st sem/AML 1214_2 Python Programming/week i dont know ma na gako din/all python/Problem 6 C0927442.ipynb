#### Write a Python class named Time that models the time of day and supports basic time arithmetic such as addition and subtraction. 

```python
#importing random library
import random

#creating a class called Time
class Time:
    
    #Initializating
    
    #init method that takes three optional parameters: hour, minute, and second, with default values of 0.
    def __init__(self,hour=0, minute=0,second=0):
        self.hour=hour
        self.minute=minute
        self.second=second
        
    #String Representation
    
    #str method to return the time formatted as a string in the form HH:MM:SS
    def __str__(self):
    
         return "{:02d}:{:02d}:{:02d}".format(self.hour, self.minute, self.second)
    
    #Addition of two time modules
    
    #add method to add two Time objects together.
    def __add__(self, other):
        total_seconds = self.hour * 3600 + self.minute * 60 + self.second +  other.hour * 3600 + other.minute * 60 + other.second
        return Time(total_seconds // 3600 % 24, total_seconds // 60 % 60, total_seconds % 60)
        
    #Subtraction of two time modules
    
    #sub method to subtract one Time object from another.
    def __sub__(self,other):
        total_seconds = self.hour * 3600 + self.minute * 60 + self.second - other.hour * 3600 + other.minute * 60 + other.second
        return Time(total_seconds // 3600 % 24, total_seconds // 60 % 60, total_seconds % 60)
 
 #Main function
#main function that takes six parameters representing the hours, minutes, and seconds of two different times
def main(h1, m1, s1, h2, m2, s2):
    #create two Time objects with these parameters
    time1 = Time(h1, m1, s1)
    time2 = Time(h2, m2, s2)
    
    #print two Time objects with these parameters
    print("Time 1:", time1)
    print("Time 2:", time2)
    
    #print the result of adding the first time to the second
    print("Time 1 + Time 2:", time1 + time2)
    #subtracting the first time from the second.
    print("Time 2 - Time 1:", time2 - time1)

# Demonstrate functionality
for i in range(10):
    print(f"  Random Time {i+1}  ")
    #random times generated by the h, m, and s functions
    h1, m1, s1 = random.randint(0, 23), random.randint(0, 59), random.randint(0, 59)
    h2, m2, s2 = random.randint(0, 23), random.randint(0, 59), random.randint(0, 59)
    main(h1, m1, s1, h2, m2, s2)
    print()


```
