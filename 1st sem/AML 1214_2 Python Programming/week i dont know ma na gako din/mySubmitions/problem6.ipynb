```python
# importing random module
import random
```

Creating a class named Time

```python
# creating a class named Time
class Time:
    """ This class takes three parameters: hour, minute and second 
    for the calculation of addition and substraction on time"""
    def __init__(self,hr=0,min=0,sec=0) -> None:
        self.hr = hr
        self.min = min
        self.sec = sec
    # defining a method to return the time formatted as a string in HH:MM:SS format
    def __str__(self) -> str:
        """ Method to return the string format"""
        return  '{:02d}:{:02d}:{:02d}'.format(self.hr,self.min,self.sec)
    
    # defining a method which adds time together
    def __add__(self, other):
        """ Method for the addition of time in correct format"""
        seconds = self.hr * 3600 + self.min*60 + self.sec
        seconds += other.hr * 3600 + other.min*60 + other.sec
        # returning the time in correct format
        return Time(seconds // 3600 % 24, seconds // 60 % 60, seconds % 60 )
    
    # defining a method which substracts time 
    def __sub__(self,other):
        """Method to subtract two times in correct format"""
        seconds = self.hr * 3600 + self.min * 60 + self.sec
        seconds -= other.hr * 3600 + other.min * 60 + other.sec
        # returning the time in correct format
        return Time(seconds // 3600 % 24, seconds // 60 % 60, seconds % 60)
     
```

```python
# creating a function to perform all the operations
def main(h1,m1,s1,h2,m2,s2):
    # creating two object of Time class 
    first = Time(h1, m1, s1)
    second = Time(h2, m2, s2)
    
    #displating the Times
    print("Time 1: ",first)
    print("Time 2: ",second)
    print("Time 1 + Time 2:", first + second)
    print("Time 1 - Time 2:", first - second)
    print()
    
def h():
    # return the hour format from 0 to 23
    return random.randint(0,23)

def m():
    # return the minute from 0 to 59
    return random.randint(0,59)

def s():
    # return the second from 0 to 59
    return random.randint(0,59)


if __name__ == "__main__":
    for _ in range(10):
        main(h(),m(),s(),h(),m(),s())
```
