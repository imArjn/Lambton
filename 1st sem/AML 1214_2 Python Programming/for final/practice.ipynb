2) Write a Python function that takes a string and counts the number of characters (character
frequency) in a string. For example, if we send 'google.com' to the function, it should return a
dictionary like this {'g': 2, 'o': 3, 'l': 1, 'e': 1, '.': 1, 'c': 1, 'm': 1}


```python
def character_frequency(word):
    frequency = {}
    for char in word:
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1
    return frequency

word = "google.com"
print(character_frequency(word))
```

```python
import string
def word_frequency_analysis(mystring):
    """ This function returns the frequency of each word in a string and also returns the count of unique alphabets in string"""
    mystring = mystring.lower()
    
    for each in mystring:
        if each in string.punctuation or each.isdigit():
            mystring = mystring.replace(each,"")
            
    print(mystring)
    
    i = 0
    start = i
    wordlist = []
    while i < len(mystring):
        if mystring[i] != ' ':
            i += 1
        else:
            wordlist.append(mystring[start:i])
            
            i += 1
            start = i
    
    if start != len(mystring):
        wordlist.append(mystring[start:])
    
    
    
    wordlist = [x for x in wordlist if x]
    print(wordlist)
    
    word_freq = {}
    for word in wordlist:
        if word not in word_freq:
            word_freq[word] = 1
        else:
            word_freq[word] += 1
    
    uniq_freq = {}
    for uniq in wordlist:
        uniq_freq[uniq] = len(set(uniq))
    return word_freq,uniq_freq
        
    
word_frequency_analysis("Hello, world! Hello, everyone.")
```

```python
import random

guess_count = 1

computer_guess = random.randint(1,20)
user_guess = int(input("Enter a number between 1 and 20: "))

while user_guess != computer_guess:
    if user_guess < computer_guess:
        print(f"Your guess {user_guess} was too low")
        user_guess = int(input(f"Enter a number greater than {user_guess}: "))
        guess_count += 1
    else:
        print(f"Your guess {user_guess} was too high")
        user_guess = int(input(f"Enter a number smaller than {user_guess}: "))
        guess_count += 1

print(f"Congratulation!!! correct answer was {computer_guess}. You guessed in {guess_count} try.")
    

```
