```python
def add_expense(clubs):
    """
    Adds an expense to a specific event within a club.
    - Prompts the user for the club name, event name, and expense amount.
    - Finds the matching club and event, then adds the expense to the event's expenses list.
    - Prints the new remaining budget for the event.
    """
    # Prompt for the club name
    clubN = input("Enter the name of the club: ")

    # Prompt for the event name
    eventN = input("Enter the name of the event: ")
    
    # Search for the club and event in the clubs list
    for each_club in clubs:
        if each_club["club_name"] == clubN:
            for each_event in each_club["events"]:
                if each_event["event_name"] == eventN:
                    # if the clubs and event are found:
                    # Prompt for the expense amount and add it to the event's expenses list
                    expense_amount = int(input("Enter the expense amount: "))
                    each_event["expenses"].append(expense_amount)
                    # Calculate and print the new remaining budget for the event
                    remaining_budget = each_event["budget"] - sum(each_event["expenses"])
                    print(f"The remaining budget for the {clubN} club and {eventN} event is {remaining_budget}")
                else:
                    print("Event is not found")    
        # Else, print a message indicating the event or club was not found
        else:
            print("Club is not found")
    


def display_budget_summary(clubs):
    """
    Displays a budget summary for each event in each club.
    - Iterates through each club and its events to print the total budget, expenses, and remaining budget.
    """
    # Iterate through each club in the clubs list
    for each_club in clubs:
        # Print the club name
        print(f"Club name is {each_club["club_name"]}")
        # Iterate through each event in the club
        for each_event in each_club["events"]:
            # Print the event name, budget, total expenses, and remaining budget
            #showing which event
            print(f"Event name is {each_event["event_name"]}")
            #showing total budget
            print(f"Total budget of the event is {each_event["budget"]}")
            # Calculate the total expenses for the event
            expenses_sum = sum(each_event["expenses"])
            print(f"Total expenses of the event is {expenses_sum}")
            # Calculate the remaining budget for the event
            remaining_budget = each_event["budget"] - expenses_sum
            print(f"The remaining budget is {remaining_budget}")


# Main program
def main():
    # Initialize the clubs data structure with sample data
    clubs = [
        {
            "club_name":"Club A",
            "events":[
                {"event_name":"event 1A","budget":5000,"expenses":[]},
                {"event_name":"event 2A","budget":3000,"expenses":[]}
            ]
            
        },
        {
           "club_name":"Club B",
            "events":[
                {"event_name":"event 1B","budget":2000,"expenses":[]},
                {"event_name":"event 2B","budget":4000,"expenses":[]}
            ] 
        }
    ]

    # Example of how to call add_expense and display_budget_summary functions
    add_expense(clubs)
    display_budget_summary(clubs)

main()
```
