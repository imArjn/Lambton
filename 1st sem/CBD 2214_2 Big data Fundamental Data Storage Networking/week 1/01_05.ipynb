```python
#Replace the NAN values with mean
for column_name,value in cars.iteritems():
    if column_name != "vichle_names":
        cars[column_name] = cars[column_name].fillna(cars[column_name].mean())
```

```python
#Replace the missing values with the most repeated value in that column
for column_name,value in cars.iteritems():
    if column_name != "vichle_names":
        cars[column_name] = cars[column_name].fillna(cars[column_name].mode())
```

```python
#Replace the NAN values with mean

#cars_filled = cars.fillna(cars.mean())
# Step 1: Identify columns to fill
columns_to_fill = [col for col in cars.columns if col != 'vichle_names']

# Step 2: Fill missing values
for column_name in columns_to_fill:
    # Fill missing values in each column with the mean of that column
    cars[column_name].fillna(cars[column_name].mean(), inplace=True)
```

```python
#Replace the NAN values with the most repeated value in that column

#cars_filled = cars.fillna(cars.mean())
# Step 1: Identify columns to fill
columns_to_fill = [col for col in cars.columns if col != 'vichle_names']

# Step 2: Fill missing values
for column_name in columns_to_fill:
    # Fill missing values in each column with the mean of that column
    cars[column_name].fillna(cars[column_name].mode(), inplace=True)
```

```python
# object_name.groupby('Series_name')
# ♔┈♔┈♔┈( WHAT THIS DOES )┈♔┈♔┈♔
# To group a  DataFrame by its values in a particular column, call the .groupby() method off of the DataFrame, and then pass
# in the column Series you want the DataFrame to be grouped by.
cars_groups = cars.groupby(cars['cyl'])
cars_groups.mean()
```
