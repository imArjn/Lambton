```python
#Avg length of banking relationship, per province

# pr_grouped = cust_df2.groupby(['Province'])
# pr_grouped.groups
# cust_df2.groupby('Province')['Len_relationship'].mean()
#for tabular form
cust_df2.groupby('Province')[['Len_relationship']].mean()
```
