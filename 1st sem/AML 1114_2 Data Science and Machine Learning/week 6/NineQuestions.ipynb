Q1. Write a function to display a Series object with only the names of Wealth customers.

```python
def wealthy_customer(dff,category):
    return dff[dff.cust_category == category]

wealthy_customer(custdf,"wealth")
```

Q2. Write a function to identify all PersonalBanking customers with more than 3 products.

```python
custdf[(custdf.num_products > 3) & (custdf.cust_category == "personal_banking")]
```

Q3. What is the average spend of customers in Alberta?

```python
def average_spend_of_X_state(df,X):
    return df[df.Province == X]["txn_total"].mean()

average_spend_of_X_state(df,"AB")
```

Q4. How many personal banking customers are there in each province?

```python
custdf[custdf.cust_category == "personal_banking"].groupby("province").size()
```

Q5. Who are the top 3 customers with the highest spend?

```python
df.sort_values("txn_total",ascending=False).head(3)["Cust_name"]
```

```python
df.groupby('Cust_name')['txn_total'].sum().sort_values(ascending=False).iloc[0:3]
```

Q6. Which customer has the highest single transaction?

```python
df[df.index == df.txn_total.idxmax()]["Cust_name"]
```

Q7. What is the average spend by Province and Customer Category?

```python
df.groupby(["Province","Cust_category"])["txn_total"].mean()
```

Q8. What is the average spend by transaction type (e.g. credit, debit, cash)?

```python
df.groupby("txn_type")["txn_total"].mean()
```

Q9. Which transaction type is the most popular in each province?

```python
df.groupby(["Province","txn_type"]).size().groupby("Province").idxmax().apply(lambda x:x[1])
```
