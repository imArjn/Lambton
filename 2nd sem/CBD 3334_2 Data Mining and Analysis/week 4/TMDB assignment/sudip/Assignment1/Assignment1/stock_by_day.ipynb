```python
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Define the ticker symbol for Comcast Corporation
ticker_symbol = "CMCSA"

# Fetch stock data for Comcast Corporation
comcast_stock = yf.Ticker(ticker_symbol)

# Define start and end dates
start_date = "2024-02-14"
end_date = "2024-03-14"

# Extract historical market data with a 1-day interval
hist = comcast_stock.history(start=start_date, end=end_date, interval="1d")

# Print the historical data
print(hist)

# Plotting the stock data
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plotting Open, Close, High, Low prices
ax1.plot(hist.index, hist['Open'], label='Open Price', color='g', marker='o', linestyle='--')
ax1.plot(hist.index, hist['Close'], label='Close Price', color='b', marker='o', linestyle='-')
ax1.plot(hist.index, hist['High'], label='High Price', color='r', marker='o', linestyle=':')
ax1.plot(hist.index, hist['Low'], label='Low Price', color='c', marker='o', linestyle='-.')

# Formatting the date on x-axis
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax1.xaxis.set_major_locator(mdates.DayLocator(interval=1))
fig.autofmt_xdate()

# Adding title and labels
ax1.set_title(f"{ticker_symbol} Stock Prices ({start_date} to {end_date}) - Daily Interval")
ax1.set_xlabel('Date')
ax1.set_ylabel('Price (USD)')
ax1.legend()
ax1.grid(True)

# Adding a secondary y-axis for volume
ax2 = ax1.twinx()
ax2.bar(hist.index, hist['Volume'], color='grey', alpha=0.3, label='Volume')
ax2.set_ylabel('Volume')
ax2.legend(loc='upper right')

plt.show()
```
