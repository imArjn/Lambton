```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the ticker symbol for Amazon
ticker_symbol = "AMZN"

# Fetch stock data for Amazon
amazon_stock = yf.Ticker(ticker_symbol)

# Define start and end dates
start_date = "2024-01-01"
end_date = "2024-06-01"

# Extract historical market data with a 1-week interval
hist = amazon_stock.history(start=start_date, end=end_date, interval="1wk")

# Print the historical data
print(hist)

# Save the data to a CSV file
hist.to_csv("amazon_stock_data_weekly.csv")

# Plotting the stock data
plt.figure(figsize=(14, 7))
plt.plot(hist.index, hist['Open'], label='Open Price', color='g')
plt.plot(hist.index, hist['Close'], label='Close Price', color='b')
plt.title(f"{ticker_symbol} Stock Prices ({start_date} to {end_date}) - Weekly Interval")
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
