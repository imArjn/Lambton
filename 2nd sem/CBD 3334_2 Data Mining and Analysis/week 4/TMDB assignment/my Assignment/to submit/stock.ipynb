# Stocks 7 Weeks before the release of "Kimetsu no Yaiba"

```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Define the ticker symbol for Sony
ticker_symbol = "SONY"

# Define the release date of the TV show (assuming it's April 6, 2019)
release_date_str = "2019-04-06"
release_date = datetime.strptime(release_date_str, "%Y-%m-%d").date()

# Calculate 7 weeks before the release date
ten_weeks_before = release_date - timedelta(weeks=7)
ten_weeks_before_str = ten_weeks_before.strftime("%Y-%m-%d")

# Calculate the end date as the day before the release date
end_date_str = (release_date - timedelta(days=1)).strftime("%Y-%m-%d")

# Fetch stock data for Sony within the calculated date range
sony_stock = yf.Ticker(ticker_symbol)
hist = sony_stock.history(start=ten_weeks_before_str, end=end_date_str, interval="1wk")

# Print the historical data
print(hist)

# Plotting the stock data
plt.figure(figsize=(14, 7))
plt.plot(hist.index, hist['Open'], label='Open Price', color='g')
plt.plot(hist.index, hist['Close'], label='Close Price', color='b')
plt.axvline(x=release_date, color='r', linestyle='--', label='TV Show Release Date')
plt.title(f"{ticker_symbol} Stock Prices (7 Weeks Before {release_date_str}) - Weekly Interval")
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

# Comparing stocks before and after the show release.

```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Define the ticker symbol for Sony
ticker_symbol = "SONY"

# Define the release date of the TV show (assuming it's April 6, 2019)
release_date_str = "2019-04-06"
release_date = datetime.strptime(release_date_str, "%Y-%m-%d").date()

# Calculate 3 weeks before the release date
three_weeks_before = release_date - timedelta(weeks=3)
three_weeks_before_str = three_weeks_before.strftime("%Y-%m-%d")

# Calculate 6 weeks after the release date
six_weeks_after = release_date + timedelta(weeks=6)
six_weeks_after_str = six_weeks_after.strftime("%Y-%m-%d")

# Fetch stock data for Sony within the calculated date range with a daily interval
sony_stock = yf.Ticker(ticker_symbol)
hist = sony_stock.history(start=three_weeks_before_str, end=six_weeks_after_str, interval="1d")

# Print the historical data (optional)
# print(hist)

# Plotting the stock data
plt.figure(figsize=(16, 8))
plt.plot(hist.index, hist['Open'], label='Open Price', marker='o', linestyle='-', color='g')
plt.plot(hist.index, hist['Close'], label='Close Price', marker='o', linestyle='-', color='b')
plt.axvline(x=release_date, color='r', linestyle='--', label='TV Show Release Date')
plt.title(f"{ticker_symbol} Stock Prices (Daily)\n3 Weeks Before and 6 Weeks After {release_date_str}")
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Showing Daily Intervals before and after

```python
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Define the ticker symbol for Comcast Corporation
ticker_symbol = "SONY"

# Fetch stock data for Comcast Corporation
comcast_stock = yf.Ticker(ticker_symbol)

# Define start and end dates
start_date = "2019-03-22"
end_date = "2019-04-22"

# Extract historical market data with a 1-day interval
hist = comcast_stock.history(start=start_date, end=end_date, interval="1d")

# Print the historical data
print(hist)

# Plotting the stock data
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plotting Open, Close, High, Low prices
ax1.plot(hist.index, hist['Open'], label='Open Price', color='g', marker='o', linestyle='--')
ax1.plot(hist.index, hist['Close'], label='Close Price', color='b', marker='o', linestyle='-')
#ax1.plot(hist.index, hist['High'], label='High Price', color='r', marker='o', linestyle=':')
#ax1.plot(hist.index, hist['Low'], label='Low Price', color='c', marker='o', linestyle='-.')
plt.axvline(x=release_date, color='r', linestyle='--', label='TV Show Release Date')

# Formatting the date on x-axis
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax1.xaxis.set_major_locator(mdates.DayLocator(interval=1))
fig.autofmt_xdate()

# Adding title and labels
ax1.set_title(f"{ticker_symbol} Stock Prices ({start_date} to {end_date}) - Daily Interval")
ax1.set_xlabel('Date')
ax1.set_ylabel('Price (USD)')
ax1.legend()
ax1.grid(True)

# Adding a secondary y-axis for volume
ax2 = ax1.twinx()
ax2.bar(hist.index, hist['Volume'], color='grey', alpha=0.3, label='Volume')
ax2.set_ylabel('Volume')
ax2.legend(loc='upper right')

plt.show()
```
