```python
# Get the shape of the image
height, width = img.shape[:2]

# Calculate the coordinates for the bottom right quarter
start_row, start_col = height // 2, width // 2

# Extract the bottom right quarter
brimg = img[start_row:, start_col:]

# Display the result
cv2.imshow('Bottom Right Quarter', brimg)
```

```python
plt.imshow(brimg,cmap="gray"),plt.axis("off")
plt.show()
```

```python
image_bgr = cv2.imread("plane2.jpg",cv2.IMREAD_COLOR)
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb),plt.axis("off")
plt.show()
```
