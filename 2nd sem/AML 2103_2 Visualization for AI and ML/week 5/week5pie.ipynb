Display 2 piecharts showing:

1) The Year-end contribution of each Product, in terms of total_units_sold
2) The contribution of each Month, towards total_year-end_profit

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('./company_sales_data.csv')

df.head()
```

```python
df.columns
```

```python
columns = df.columns
```

sol 1

```python
plt.figure(figsize=(10,10))
plt.pie(x=df.sum(), autopct='%1.1f%%' , labels=columns)
plt.title("Total sales of all the products throughtout the year")
plt.show()
```

sol 2

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul","Aug", "Sep", "Oct", "Nov", "Dec"]

plt.figure(figsize=(10,10))
plt.pie(x = df['total_profit'], autopct='%1.1f%%', labels=months)
plt.title("Contribution of each Month in total sales throughtout the year")
plt.show()
```
