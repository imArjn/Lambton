## Word Cloud -> 13 July

```python
import wordcloud
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
stop_words = list(STOPWORDS)[:]
print(stop_words)
```

```python
stop_words = set(stop_words)
```

```python
# Define the URL of the Gist file
url = "https://gist.githubusercontent.com/phillipj/4944029/raw/75ba2243dd5ec2875f629bf5d79f6c1e4b5a8b46/alice_in_wonderland.txt"

# Fetch the content from the URL
response = urllib.request.urlopen(url)
data = response.read().decode('utf-8')

# Print the fetched content (you can process it as needed)
print(data[:1000])
```

```python
words = data.split()
```

```python
filtered_words = [w for w in words if w.lower() not in stop_words]
```

```python
filtered_text = " ".join(filtered_words)
```

```python
print(filtered_text[:1000])
```

```python
filtered_text_1000 = filtered_text[:1000]
```

```python
wordcloud = WordCloud(stopwords=stop_words, collocations=False).generate(filtered_text)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Add "said" in stopwords
stop = list(stop_words) + ["said"]
```

```python
stop_words = set(stop)
print(len(stop_words))
```

```python
wordcloud = WordCloud(stopwords=stop_words, collocations=True, background_color="White").generate(filtered_text)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
