For the above graph, annotations every point with its corresponding Txn Amount:

plt.annotate()

```python
plt.figure(figsize=(15,5))
plt.plot(df['txn_id'], df['txn_total'], 'r--o', label = "TRANSACTION_DOLORS", markerfacecolor='blue')
plt.plot(df['txn_id'], df['txn_total'], color = "c")
plt.title("Line graph of Txn Amounts")
plt.xlabel("Txn_ID")
plt.ylabel("Txn_amount in $", rotation = 0, labelpad=40)
plt.legend()

# Annotate each point with its corresponding transaction amount
for i, txt in enumerate(df['txn_total']):
    plt.annotate(txt, (df['txn_id'][i], df['txn_total'][i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```

Alternative Way:

```python
plt.figure(figsize=(15,5))
plt.plot(df['txn_id'], df['txn_total'], 'r--o', label = "TRANSACTION_DOLORS", markerfacecolor='blue')
plt.plot(df['txn_id'], df['txn_total'], color = "c")
plt.title("Line graph of Txn Amounts")
plt.xlabel("Txn_ID")
plt.ylabel("Txn_amount in $", rotation = 0, labelpad=40)
plt.legend()

# Annotate each point with its corresponding transaction amount
for x,y in zip(df['txn_id'], df['txn_total']):
    label = "{:.2f}".format(y)
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(0, 10), ha='center')
```

# plt.bar()

For `plt.bar()` : `plt.bar(x,heights, width)`

- x = list of x coordinates where the bars are centered
- heights = list of heights of each bar
- width = a single value; width of each bar; default = 0.8

```python
plt.bar(["A","B","C","D"],[40,80,15,10],width = 0.5)
```

### Grouped bar graph:<br>
a) For a vertical bar graph: Each Category on the x axos, has multiple subcategories<br>
b) For a horizontal bar graph: Each Category on the y-axis, has multiple subcategories

`plt.barh()` : horizontal bar-graph

```python
plt.barh(["A","B","C","D"],[40,80,15,10],height= 0.5)
```

```python
courses_list = ["CSE","EXTC","Mech","Aero","Mining"]

barIndexes_alongXaxis = np.arange(len(courses_list))

girlsCount_barHeights = [150,100,50,20,8]
boysCount_barHeights = [100,150,100,53,27]

w=0.4

bar1 = barIndexes_alongXaxis # centers 0,1,2,3,4
bar2 = [(i+w) for i in bar1]

plt.xticks((bar1+w/2),courses_list)

plt.bar(bar1,girlsCount_barHeights, width = w,label = "Girls")
plt.bar(bar2,boysCount_barHeights, width = w,label = "Boys")

plt.legend()

```
